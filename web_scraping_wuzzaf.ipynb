```python
Job_Title = []
Company_Name = []
Job_Type = []
Job_Location = []
```

```python
import pandas as pd
import csv
```

```python
for i in range(job_number):
    job_card = job_post[i]
    Title = job_card.find("h2").text.strip()
    Location = job_card.find("span",class_="css-16x61xq").text.strip()
    Type = job_card.find("span",class_="css-uc9rga eoyjyou0").text.strip()
    Company = job_card.find("a",class_="css-ipsyv7").text.strip()

    Job_Title.append(Title)
    Company_Name.append(Company)
    Job_Type.append(Type)
    Job_Location.append(Location)

df=pd.DataFrame({"Job Title":Job_Title,
                     "Company Name":Company_Name,
                     "Job Type":Job_Type,
                     "Job Location":Job_Location})
df.to_csv("Wuzzaf.csv", encoding="utf-8")
```
